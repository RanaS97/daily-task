```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
df=pd.read_csv('/content/insurence.csv')
```

```python
sns.lmplot(x='age',y='premium',data=df)
```

```python
lr=linear_model=LinearRegression()
lr.fit(df[['age']].values,df['premium'].values)
```

```python
lr.predict([[50]])
```

```python
lr.coef_
```

```python
lr.intercept_
```
